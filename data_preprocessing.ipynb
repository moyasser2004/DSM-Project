### use bar to show which store has maximum sales

```python
grouped_data = data_frame.groupby("Store")["Weekly_Sales"].sum()

store_ids = grouped_data.index
total_sales = grouped_data.values

plt.figure(figsize=(10, 6))
bars = plt.bar(store_ids, total_sales, color="skyblue")
plt.xlabel("Store")
plt.ylabel("Total Weekly Sales")
plt.title("Total Weekly Sales per Store")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()

for bar, store_id in zip(bars, store_ids):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        str(store_id),
        ha="center",
        va="bottom",
        fontsize=8,
        color="black",
    )

plt.show()
```

```python
store_has_max_sales = data_frame.groupby("Store")["Weekly_Sales"].sum().idxmax()
print("Store with maximum sales:", store_has_max_sales)
```

### get Store with maximum standard deviation:

```python
store_has_max_std = data_frame.groupby("Store")["Weekly_Sales"].std().idxmax()
print("Store with maximum standard deviation:", store_has_max_std)
```

### get Holidays with higher sales than mean non-holiday sales

```python
mean_non_holiday_sales = data_frame[data_frame["Holiday_Flag"] == 0][
    "Weekly_Sales"
].mean()


high_sales_holidays = data_frame[
    (data_frame["Holiday_Flag"] == 1)
    & (data_frame["Weekly_Sales"] > mean_non_holiday_sales)
]

print(
    "Holidays with higher sales than mean non-holiday sales:",
    high_sales_holidays["Date"].unique(),
)
```

### show bar to explane the Holidays with higher sales than mean non-holiday sales

```python
unique_high_sales_dates = high_sales_holidays["Date"].unique()
plt.figure(figsize=(12, 6))
plt.bar("Mean Non-Holiday Sales", mean_non_holiday_sales, color="skyblue")

for idx, date in enumerate(unique_high_sales_dates):
    total_sales = high_sales_holidays[high_sales_holidays["Date"] == date][
        "Weekly_Sales"
    ].sum()
    plt.bar(
        idx + 1,
        total_sales,
        color=plt.cm.tab10(idx / len(unique_high_sales_dates)),
        label=date,
    )

plt.xlabel("Date")
plt.ylabel("Total Weekly Sales")
plt.title(
    "Total Weekly Sales for Holidays with Higher Sales than Mean Non-Holiday Sales"
)

plt.legend()
plt.tight_layout()
plt.show()
```

### get monthly sales & semesters sales
- #### for tow semester sales as semester contain 6 months

```python
data_frame["Month"] = data_frame["Date"].dt.month
monthly_sales = data_frame.groupby("Month")["Weekly_Sales"].sum()

data_frame["Semester"] = data_frame["Date"].dt.quarter.map({1: "Q1", 3: "Q2"})
semester_sales = data_frame.groupby("Semester")["Weekly_Sales"].sum()


print("Monthly Sales:\n", monthly_sales)

print("\nSemester Sales:\n", semester_sales)
```

### get monthly sales & semesters sales
- #### for four semester sales as semester contain 4 months

```python
data_frame["Month"] = data_frame["Date"].dt.month
monthly_sales = data_frame.groupby("Month")["Weekly_Sales"].sum()


data_frame["Semester"] = data_frame["Date"].dt.quarter.map(
    {
        1: "Q1",
        2: "Q2",
        3: "Q3",
        4: "Q4",
    }
)
semester_sales2 = data_frame.groupby("Semester")["Weekly_Sales"].sum()


print("Monthly Sales:\n", monthly_sales)
print("\nSemester Sales:\n", semester_sales2)
```

### show plot for relation between weekly sales and "Temperature", "Fuel_Price", "CPI", "Unemployment"

```python
plt.figure(figsize=(12, 8))
numeric_features = ["Temperature", "Fuel_Price", "CPI", "Unemployment"]


for feature in numeric_features:
    plt.subplot(2, 2, numeric_features.index(feature) + 1)
    sns.scatterplot(x=feature, y="Weekly_Sales", data=data_frame)
    plt.title(f"Weekly Sales vs {feature}")
    plt.xlabel(feature)
    plt.ylabel("Weekly Sales")

plt.tight_layout()
plt.show()
```

### show histogram

- #### The histogram clearly shows the range of Temperature. may be averge position
- #### The histogram clearly shows the range of Unemployment. skewed to 6.5 to 9
- #### The histogram clearly shows the range of CPI. skewed different positions

```python
plt.figure(figsize=(12, 8))
numeric_features = ["Temperature", "Fuel_Price", "CPI", "Unemployment"]

for feature in numeric_features:
    plt.subplot(2, 2, numeric_features.index(feature) + 1)
    sns.histplot(data_frame[feature], kde=True, bins=15, color="skyblue")
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
